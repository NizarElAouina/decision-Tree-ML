<h1> Q2:
    <h3>Il ya des lignes avec des données manquantes et la collone DEBTINC contient beaucoup de données manquantes. Donc on a procédé à la suppression des lignes et colonnes qui présentent un problème

```python
def Entropie(S):
    classes_distinctes = set(S['BAD'])
    classes = list(S['BAD'])
    s=0
    for c in classes_distinctes:
        p = classes.count(c)/len(classes)
        s+=p*log2(p)
    return (-1.0*s)
print(Entropie(data))
data
```

    Ce résultat signifie que les valeurs dans cette colonne sont relativement dispersées et présentent un certain niveau d'incertitude ou de désordre. 

<h1> Q3 : Indice de GINI et gain d'informations

```python
def Gain_d_information(S, attribut):
    valeurs_attribut = set(S[attribut])
    somme = 0
    for v in valeurs_attribut:
        Sv = S[S[attribut] == v]
        somme += (len(Sv) / len(S)) * Entropie(Sv)
    return Entropie(S) - somme

for col in data.columns:
    if col != 'BAD':
        print(col + ':', Gain_d_information(data, col))
```

    La gain d'information reflète l'importance de chaque feature dans la classification.
    Dans notre cas, il sert à détecter les features les plus influentes dans l'octroi des crédits.
    De ce fait on relarque que les features MORTDUE, VALUE et CLAGE sont celles avec le plus grand gain d'informations et donc les plus décisives dans l'octroi de crédits.
        Tandis qu'on trouve des features avec un gain d'information assez faible comme la feature LOAN, voire très faible comme la feature REASON.



```python
def GINI(S):
    classes_distinctes = set(S['BAD'])
    classes = list(S['BAD'])
    total_instances = len(classes)
    gini = 1.0
    for c in classes_distinctes:
        p = classes.count(c) / total_instances
        gini -= p ** 2
    return gini
GINI(data)
```

    L'indice de Gini proche de 0 indique que l'ensemble de données est très pur, c'est-à-dire que toutes les étiquettes de classe sont les mêmes. Un indice de Gini proche de 1 indique que l'ensemble de données est très impur, c'est-à-dire que les étiquettes de classe sont réparties de manière égale ou presque égale.
    Dans notre cas, nous avons une valeur égale à 0.3173342099014311; ceci reflète une impureté relativement moyenne, contrairement à l'entropie qui avait révélé un désordre plus important.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split

target=['BAD']
feat=['LOAN','MORTDUE','VALUE','YOJ','DEROG','DELINQ','CLAGE','NINQ','CLNO']
training_data, testing_data = train_test_split(data, test_size=0.2, random_state=1)

mdl_gini = DecisionTreeClassifier(criterion ='gini')
mdl_entro = DecisionTreeClassifier(criterion ='entropy')
mdl_gini.fit(training_data[feat],training_data[target])
mdl_entro.fit(training_data[feat],training_data[target])
```

<H1>Q6

```python
print("Indice de GINI:")
y_pred_gini = mdl_gini.predict(testing_data[feat])
print("Matrice de confusion: ",)
print(confusion_matrix(testing_data[target], y_pred_gini))
print ("Accuracy : ",accuracy_score(testing_data[target],y_pred_gini)*100)

print("REntropie:")
y_pred_entropy = mdl_gini.predict(testing_data[feat])
print("Matrice de confusion: ",)
print(confusion_matrix(testing_data[target], y_pred_entropy))
print ("Accuracy : ",accuracy_score(testing_data[target],y_pred_entropy)*100)
```

    Nous remarquons que les résultats donnés par les deux modèle sont identiques.
    La précision du modèle est de 90.35%, ce qui en fait un modèle satisfaisant.
    Nous notons par ailleurs que la matrice de confusion a un nombre de faux négatifs environ 3 fois plus important que les faux positifs.
