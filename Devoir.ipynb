<h1> Q2:
    <h3>Il ya des lignes avec des donnÃ©es manquantes et la collone DEBTINC contient beaucoup de donnÃ©es manquantes. Donc on a procÃ©dÃ© Ã  la suppression des lignes et colonnes qui prÃ©sentent un problÃ¨me

```python
def Entropie(S):
    classes_distinctes = set(S['BAD'])
    classes = list(S['BAD'])
    s=0
    for c in classes_distinctes:
        p = classes.count(c)/len(classes)
        s+=p*log2(p)
    return (-1.0*s)
print(Entropie(data))
data
```

    Ce rÃ©sultat signifie que les valeurs dans cette colonne sont relativement dispersÃ©es et prÃ©sentent un certain niveau d'incertitude ou de dÃ©sordre. 

<h1> Q3 : Indice de GINI et gain d'informations

```python
def Gain_d_information(S, attribut):
    valeurs_attribut = set(S[attribut])
    somme = 0
    for v in valeurs_attribut:
        Sv = S[S[attribut] == v]
        somme += (len(Sv) / len(S)) * Entropie(Sv)
    return Entropie(S) - somme

for col in data.columns:
    if col != 'BAD':
        print(col + ':', Gain_d_information(data, col))
```

    La gain d'information reflÃ¨te l'importance de chaque feature dans la classification.
    Dans notre cas, il sert Ã  dÃ©tecter les features les plus influentes dans l'octroi des crÃ©dits.
    De ce fait on relarque que les features MORTDUE, VALUE et CLAGE sont celles avec le plus grand gain d'informations et donc les plus dÃ©cisives dans l'octroi de crÃ©dits.
        Tandis qu'on trouve des features avec un gain d'information assez faible comme la feature LOAN, voire trÃ¨s faible comme la feature REASON.



```python
def GINI(S):
    classes_distinctes = set(S['BAD'])
    classes = list(S['BAD'])
    total_instances = len(classes)
    gini = 1.0
    for c in classes_distinctes:
        p = classes.count(c) / total_instances
        gini -= p ** 2
    return gini
GINI(data)
```

    L'indice de Gini proche de 0 indique que l'ensemble de donnÃ©es est trÃ¨s pur, c'est-Ã -dire que toutes les Ã©tiquettes de classe sont les mÃªmes. Un indice de Gini proche de 1 indique que l'ensemble de donnÃ©es est trÃ¨s impur, c'est-Ã -dire que les Ã©tiquettes de classe sont rÃ©parties de maniÃ¨re Ã©gale ou presque Ã©gale.
    Dans notre cas, nous avons une valeur Ã©gale Ã  0.3173342099014311; ceci reflÃ¨te une impuretÃ© relativement moyenne, contrairement Ã  l'entropie qui avait rÃ©vÃ©lÃ© un dÃ©sordre plus important.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split

target=['BAD']
feat=['LOAN','MORTDUE','VALUE','YOJ','DEROG','DELINQ','CLAGE','NINQ','CLNO']
training_data, testing_data = train_test_split(data, test_size=0.2, random_state=1)

mdl_gini = DecisionTreeClassifier(criterion ='gini')
mdl_entro = DecisionTreeClassifier(criterion ='entropy')
mdl_gini.fit(training_data[feat],training_data[target])
mdl_entro.fit(training_data[feat],training_data[target])
```

```python
print("Indice de GINI:")
y_pred_gini = mdl_gini.predict(testing_data[feat])
print("Matrice de confusion: ",)
print(confusion_matrix(testing_data[target], y_pred_gini))
print ("Accuracy : ",accuracy_score(testing_data[target],y_pred_gini)*100)

print("REntropie:")
y_pred_entropy = mdl_gini.predict(testing_data[feat])
print("Matrice de confusion: ",)
print(confusion_matrix(testing_data[target], y_pred_entropy))
print ("Accuracy : ",accuracy_score(testing_data[target],y_pred_entropy)*100)
```
